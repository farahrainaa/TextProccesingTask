```python
from bs4 import BeautifulSoup
```

```python
import requests
page = requests.get('https://kprofiles.com/kdrama-list/')
bs = BeautifulSoup(page.text, 'html.parser')
bs
```

```python
bs.find_all('a', {'rel':'noopener'})
```

```python
list_drama = bs.find_all('a', {'rel':'noopener'})
for li in list_drama:
    print(li.text)
```

```python
korean_drama = []

for li in list_drama:
    drama = li.text.strip()

    if drama not in ['Groups Debut Dates', 'Current Kpop Pre-Debut Groups'] and drama not in korean_drama:
       korean_drama.append(drama)

sorted_drama = sorted(korean_drama)
```

```python
print("Berikut adalah list Korean Drama:")

for drama in sorted_drama:
    print(drama)
```
